```python
#Percentage of matches won

win_percent=(b5_df['W']/b5_df['MP'])*100
print(win_percent)
win_percent.plot(kind='bar', figsize=(50,40))
plt.ylabel=("win_percent")
plt.xlabel=("Squad")
```

```python
#Percentage of matches dropped

drop_percent=((b5_df['MP']-b5_df['W'])/b5_df['MP'])*100
print(drop_percent)
drop_percent.plot(kind='bar', figsize=(50,40))
plt.ylabel=("drop_percent")
plt.xlabel=("Squad")
```

```python
#Pie chart for avg matches won in different country

country_w =b5_df.groupby('Country')['W'].mean().sort_values()
print(country_w)
country_w.sort_values().plot(kind='pie', figsize=(10,10))
plt.ylabel=("country_w")
plt.xlabel=("Country")
```

```python
#Total goals scored

total_gs = b5_df['GF'].sum()
print(total_gs)
```

```python
#Goals Scored - Country-wise

country_gs =b5_df.groupby('Country')['GF'].sum().sort_values()
print(country_gs)
country_gs.sort_values().plot(kind='bar', figsize=(5,5))
plt.ylabel=("country_gs")
plt.xlabel=("Country")
```

```python
#Top Scorer contribution

ts_contri=(b5_df['Goals By Top Team Scorer']/b5_df['GF'])*100
ts_contri.plot(kind='bar', figsize=(30,20))
```

```python
#Most Valuable Players to the team


ts_contri.nlargest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
#Goal concede by goalkeepers

gk_concede=b5_df['GA']
print(gk_concede)
gk_concede.plot(kind='bar', figsize=(50,40))
plt.ylabel=("gk_concede")
plt.xlabel=("Goalkeepers")
```

```python
#Goals concede country-wise

country_gk =b5_df.groupby('Country')['GA'].sum().sort_values()
print(country_gk)
country_gk.sort_values().plot(kind='bar', figsize=(5,5))
plt.ylabel=("country_gk")
plt.xlabel=("Country")
```

```python
#Trend of points per game vs ranks

plt.plot(b5_df['Pts/G'])
plt.show()
```

```python
#Goal difference and Expected goal difference

plt.plot(b5_df['GD'])
plt.show()
plt.plot(b5_df['xGD'])
plt.show()
```

```python
#Avg Attendance for the season

avg_attend=b5_df['Avg Attendance'].mean()
print(avg_attend)
```

```python
#Attendance country-wise

country_attend =b5_df.groupby('Country')['Avg Attendance'].sum().sort_values()
print(country_attend)
country_attend.sort_values().plot(kind='bar', figsize=(5,5))
plt.ylabel=("country_attend")
plt.xlabel=("Country")
```
